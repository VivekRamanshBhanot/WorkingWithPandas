```python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
print(pop.head())
print(areas.head())
print(abbrevs.head())
```

```python
x = pd.merge(pop,abbrevs,left_on='state/region',right_on='abbreviation',how='outer')
print(x.head())
x = x.drop('abbreviation',1)
print('\n',x.head())
```

```python
x.isnull().any()
```

```python
x[x['population'].isnull()].head()
```

```python
x[x['state'].isnull()]
```

```python
x.loc[x['state'].isnull(),'state/region'].unique()
```

```python
x.loc[x['state/region'] == 'PR','state'] = 'Puerto Rico'
x.loc[x['state/region'] == 'USA','state'] = 'United States'
```

```python
x.isnull().any()
```

```python
print(areas[areas['area (sq. mi)'].isnull()])
final = pd.merge(x,areas,how='left',on='state')
final.loc[final['area (sq. mi)'].isnull(),'state'].unique()
```

```python
q =final.loc[final['area (sq. mi)'].isnull(),'area (sq. mi)'] = final.loc[final['area (sq. mi)'].notnull(),'area (sq. mi)'].unique().sum()
print(q)
print(final.loc[final['state'] == 'United States','area (sq. mi)'].unique())
```

```python
final.dropna(inplace=True)
```

```python
data2010 = final[(final['year'] == 2010) & (final['ages'] == 'total')]
data2010.head()
```

```python
data2010.set_index('state',inplace=True)
data2010.head()
```

```python
density = data2010['population']/data2010['area (sq. mi)']
density.sort_values(ascending=False, inplace=True)
print(density.head())
print(density.tail())
```

```python
df = pd.DataFrame({'key':list('ABCABC'),
                  'data1':range(6),
                  'data2':np.random.randint(0,10,6)},columns=['key','data1','data2'])
print(df)
print(df.groupby('key').aggregate(['std','min','max']))
print(df.groupby('key').aggregate({'data1':'min','data2':'max'}))
```

```python
def p(x):
    if x['data2'].sum() != 9:
        x['data2']=100
        print(x)
    return x
df.groupby('key').apply(p)
```

```python
def p(x):
    y = False
    if (x['method'] == 'Pulsar Timing').any():
        y = True
    return y
planets.groupby('method').filter(p)
```
